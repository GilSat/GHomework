```python
#Future Value Calc
principal = float(input("What is the future value?")) 


#find out what the rate is and do data validationto ensure it is a percent
rate = float(input("What rate do you expect to earn yearly?"))
if(rate > 1):
  rate /= 100

#find out how long this will be for 
time = int(input("How many years to you expect to hold this investment?"))

output = principal/(1+rate)**time
print(f"The present value of your inmvestment is  ${output:,.2f}")
```
