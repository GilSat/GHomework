```python
import numpy as np

#avg of two numbers
def avg_of_2(x,y):
  return (x+y)/2

#avg of a list of numbers
def avg_of_list(myNums):
  total = 0
  for i in myNums:
    total = total + i
  return total/len(myNums)

#getting the weighted average of a np.array
def expectedValue1(info, weights):
  return (info @ weights)



num_one = 2
num_two = 5
print(avg_of_2(num_one,num_two))

z = [1,2,3,4,5,6,7,8]
print(avg_of_list(z))

info = np.array([2,3,4,5,6])
weights = np.array([.1,.15,.07,.32,.36])
print(expectedValue1(info,weights))
```

```python
#sum from a to b
def sumRange(a,b):
  total = 0
  for i in range(a,b+1):
    total = total + i
  return total

print(sumRange(5,8))
```

```python
#Takes in cost and discount and tax gives total
def finalCost(item_cost,discount,tax_rate):
  discount = item_cost * (discount / 100)
  item_cost = item_cost - discount
  item_cost = item_cost * (1+tax_rate)
  return item_cost

tax = .08625
itemcost = 100
percentoff = 20
print(finalCost(itemcost,percentoff,tax))
```

```python
#list of primes below some n
#first create a function that tells you if a number is prime

def isPrime(number):
  for i in range(2,number):
    if(number%i == 0):
      return False
  return True

def listPrimes(number):
  for i in range(2,number+1):
    if(isPrime(i)):
      print(i, end=" ")

listPrimes(100)
```

```python
#list all numbers below some number that is coprime to it
#coprime means that a and b only share 1 as a divisor
#for instance 12 shares only a 1 with 1,5,7 and 11
#the key here is to break down the problem into multiple functions
def isCoprime(a,b):
  if(a>b):
    holder = a
    a = b
    b = holder
  
  for i in range(2,a+1):
    if(a%i == 0 and b%i == 0):
      return False
  return True

def listCoprimes(number):
  for i in range(1, number):
    if(isCoprime(i, number)):
      print(i, end=" ")

listCoprimes(12)
```

```python
#list twin primes

def isTwinPrime(s):
  if(isPrime(s) and isPrime(s+2)):
    return True
  return False

def listTwinPrimes(number):
  for i in range(2,number+1):
    if(isTwinPrime(i)):
      print(i,i+2)

listTwinPrimes(100)
```

```python
#testing rock paper scissors
import random as rnd 

rock = paper = scissors = 0
for i in range(100000):
  playerOne = rnd.randint(1,3)
  playerTwo = rnd.randint(1,3)

  if ((playerOne == 1 and playerTwo == 3) or (playerTwo == 1 and playerOne == 3)):
    rock = rock + 1
  elif((playerOne == 2 and playerTwo == 1) or (playerTwo == 2 and playerOne == 1)):
    paper = paper + 1
  elif((playerOne == 3 and playerTwo == 2) or (playerTwo == 3 and playerOne == 2)):
    scissors = scissors + 1

print(rock/100000)
print(paper/100000)
print(scissors/100000)
print((100000-(rock+paper+scissors))/100000)
```
