```python
#load pandas and numpy libraries 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import missingno as msno
import seaborn as sns

df = pd.read_csv("https://raw.githubusercontent.com/Opensourcefordatascience/Data-sets/master/admission.csv")
df.head()

```

```python
#check for nulls
df.isnull().sum()
```

```python
#stats on the data
df.nunique()
```

```python
df.describe()
```

```python
sns.catplot(data=df , kind='box' , height=4, aspect=2)
```

```python
sns.catplot(x='gre', data=df , kind='count', height=3, aspect=3)
```

```python
df.hist(figsize=(15,10), bins=20, xlabelsize=10, ylabelsize=10)
```

```python
df.plot(x='gpa', y = 'gre', kind='scatter')
plt.show()
```

```python
sns.distplot(df['gpa'])
```

```python
corr = df.corr()
fig, ax = plt.subplots(figsize=(10,10))
sns.heatmap(data=corr, square=True, annot=True, cbar=True, ax = ax)
```
