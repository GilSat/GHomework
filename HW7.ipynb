1. 
Create functions to do the following: max, min, average, standard deviation, and geometric average. 

```python
import numpy as np
import math as m

def theMax(numbers):
  my_max = numbers[0]
  for i in numbers:
    if my_max < i:
      my_max = i
  return my_max

def theMin(numbers):
  my_min = numbers[0]
  for i in numbers:
    if my_min > i:
      my_min = i
  return my_min

def theAverage(numbers):
  total = 0 
  for i in numbers:
    total += i
  return total/len(numbers)

def theGeoMean(numbers):
  total = 1
  for i in numbers:
    total *= i
  return total**(1/len(numbers))

def theStandardDev(numbers):
  theAvg = theAverage(numbers)
  total = 0
  for i in numbers:
    total = total + (i-theAvg)**2
  return m.sqrt(total/(len(numbers)-1))


a = [8,6,2,9,13,3]
print(theMax(a))
print(theGeoMean(a))
print(theStandardDev(a))
```

```python
#create an empty list
def concatPositions(list1, list2):
  myBigList = []
  for i in range(len(list1)):
    myBigList.append(float(str(list1[i])+str(list2[i])))
  return myBigList


list1 = [1,2,3]
list2 = [7,8,9]
bigList = concatPositions(list1,list2)
print(bigList)
```

5.Create a function that returns a list of primes up until the number passed in. So if 100 is passed in, it should return all the primes up until 100.

```python
def isPrime(num):
  for i in range(2,num):
    if (num%i == 0):
      return False
  return True

def listPrimes(amount):
  #an empty list
  myList = []
  count = 0 
  checkVal = 2
  while (count < amount):
    if(isPrime(checkVal)):
      myList.append(checkVal)
      count += 1
    checkVal += 1
  return myList

print(listPrimes(100))
```

8.Find the sum of the even valued Fibonacci numbers below 500.


```python
bound = 500
numOne = 1
numTwo = 1
numThree = numOne + numTwo
total = 0

while numThree < bound:
  numThree = numOne + numTwo
  if numThree%2 == 00 and numThree < bound:
    total += numThree
  numOne = numTwo
  numTwo = numThree

print(total)
```
