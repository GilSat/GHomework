##HW4

```python
item_price = 99
amount = 0
discount = 0
total_price = 0

#ask the user for the number ordered
amount = int(input("How many would you like to purchase? "))

if (amount >= 10 and amount <= 19):
  discount = .1
elif (amount >= 20 and amount <= 49):
  discount = .2
elif (amount >= 50 and amount <= 99):
  discount = .3
elif (amount >= 100):
  discount = .4

print (f"The discount is {discount*100:.0f}% off")

#calculate and print total price
total_price = amount*item_price*(1-discount)
print(f"The total price is ${total_price:,.2f}")
```

## HW 4 - Question 2

```python
weight = 0
rate = 0.0
total_price = 0

weight = float(input("How heavy it the package? "))

if (weight <= 2):
  rate = 1.5
elif (weight > 2 and weight <= 6):
  rate = 3.0
elif (weight > 6 and weight <= 10):
  rate = 4.0
elif (weight > 10):
  rate = 4.75

print (f"Your shipping rate is ${rate:,.2f}")

total_price = rate*weight
print (f"The price to ship is: ${total_price:,.2f}")

```
